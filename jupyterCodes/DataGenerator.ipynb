## 2. Helpers for our Helpers

```python
def uniform(low, up, size=None, seed=None):
    '''
    Intro:
        helper function that returns one list of uniform random values.
    ---
    Inputs: 
        low:
            The lower limit of our possible values.
        up:
            The uper limit of our possible values.
        size:
            How many points we want.
    ---
    Outputs:
        A list size values between [low, up].
    '''   
    random.seed(seed)
    try:
        return [random.uniform(a, b) for a, b in zip(low, up)]
    except TypeError:
        return [random.uniform(a, b) for a, b in zip([low] * size, [up] * size)]
```

## 3. Creating Random Datasets

```python
def createRandomDataForZDT1(n, seed=None):
    '''
    Intro:
        This data comes from a Gaussian uniform variation
        for the function deap.benchmarks.zdt1
    ---
    Inputs: 
        n:
            is the number of points that we want to receive.
    ---
    Outputs:
        A list of 100 points of dimension 30 with values between [0, 1].
        Each point is a list.
    '''   
    random.seed(seed)
    BOUND_LOW, BOUND_UP = 0.0, 1.0
    NDIM = 30

    result_data = []
    for i in range(n):
        if seed is None:
            temp_list = uniform(BOUND_LOW,BOUND_UP,30,None)
        else:
            temp_list = uniform(BOUND_LOW,BOUND_UP,30,seed*i)
        result_data += [temp_list]
    
    return result_data
```
